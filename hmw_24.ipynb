```python
from PIL import Image
import matplotlib.pyplot as plt

photo = Image.open(r'C:\Users\User\Downloads\jinny.jpg')

rotated_photo = photo.rotate(120, expand=True)

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(photo)
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(rotated_photo)
axes[1].set_title("Rotated Image (120)")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

```python
np.random.seed(55)
arr1 = np.random.randint(1, 101, (4, 4))
col_sum = np.sum(arr1, axis=0)
col_max = np.max(arr1, axis=0)
col_min = np.min(arr1, axis=0)
col_avg = np.mean(arr1, axis=0)

row_sum = np.sum(arr1, axis=1)
row_max = np.max(arr1, axis=1)
row_min = np.min(arr1, axis=0)
row_avg = np.mean(arr1, axis=0)

print(f'{arr1}\n')
print('each column stats:')
print(f'col_sum {col_sum}, col_max {col_max}, col_min {col_min}, col_avg {col_avg}\n')
print('each row stats:')
print(f'row_sum {row_sum}, row_max {row_max}, row_min {row_min}, row_avg {row_avg}\n')
print(arr1[arr1>50])
```
